````python
from langgraph.graph import StateGraph, START, END
from langchain_openai import ChatOpenAI
from typing import TypedDict
from dotenv import load_dotenv
from langchain_groq import ChatGroq
import os
from pydantic import BaseModel, Field
````

````python
load_dotenv()
````

````python
model = ChatGroq(
    model="llama-3.1-8b-instant",
    temperature=0,
    response_format={"type": "json_object"}
)
````

When expecting a structured output, always use this model of open ai:
```
model = ChatOpenAI(model = 'gpt-4o-mini')

```
- Then we create a Schema which consist of:
   - Outputs we are expecting from model
   - Using Pydantic BaseModel and Field.
   - Syntax of Field:
   ```
   Field(description = "Whatever you want to write")
   Field(description = "Whatever", ge = 0, le = 10) #ge = greater than equals to | le = less than equals to
   ```
   - We can specify range also in ```Field()```


````python
class Evaluation_Schema(BaseModel):
    feedback: str = Field(description="Detailed feedback for essay. Do NOT include score here")
    score: int = Field(description="Score out of 10", ge =0, le = 10)
````

Then using a function ```model.with_structured_output()``` we will passs our Schema in here

````python
struct_model = model.with_structured_output(Evaluation_Schema)
````

````python
prompt = f"""Evaluate the language quality of the following essay and provide a feedback and assign a score out of 10

STRICT RULES:
- Return ONLY valid JSON
- Do NOT include score inside feedback
- score must be a separate integer field

Format:
{{
  "feedback": "...",
  "score": ..
}}


 \n {essay}"""

struct_model.invoke(prompt)
````
